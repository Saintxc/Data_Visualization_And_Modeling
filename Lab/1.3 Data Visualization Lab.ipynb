```python
import pandas as pd

df = pd.read_csv('spotify-2023.csv', encoding='ISO-8859-1')

df.head()
```

```python
df['bpm'].plot(kind ='hist', title="BPM Distribution", edgecolor = 'black')
```

```python
df['key'].value_counts().plot(kind ='bar', title="Key Distribution", edgecolor = 'black')
```

```python
df['mode'].value_counts().plot(kind ='bar', title="Mode Distribution", edgecolor = 'black')
```

```python
df['danceability_%'].plot(kind ='hist', title="Dancebility Distribution", edgecolor = 'black')
```

```python
df['energy_%'].plot(kind ='hist', title="Energy Distribution", edgecolor = 'black')
```

```python
df['speechiness_%'].plot(kind ='hist', title="Speechiness Distribution", edgecolor = 'black')
```

```python
df['acousticness_%'].plot(kind ='hist', title="Acousticness Distribution", edgecolor = 'black')
```

```python

filtered_df = df [
    (df['bpm'] >= 120) & (df['bpm'] <= 140) & 
    (df['key'] == 'C#') & 
    (df['mode'] == 'Major') *
    (df['danceability_%'] >= 70) &
    (df['energy_%'] >= 60) & (df['energy_%'] <= 80) *
    (df['speechiness_%'] <= 10) &
    (df['acousticness_%'] <= 20)
]

filtered_df.head()
```
